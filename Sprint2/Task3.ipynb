Описание задачи:

Создать базовый класс CreditCard в C#, который будет представлять информацию о кредитных картах. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма.

Требования к базовому классу CreditCard:

• Атрибуты: Номер карты (CardNumber), Холдера (HolderName), Срок
действия (ExpiryDate).

• Методы:

o GetInfo(): метод для получения информации о кредитной карте в виде
строки.
o Pay(): метод для оплаты покупки с использованием карты.
o CheckBalance(): метод для проверки баланса на карте.

Требования к производным классам:
1. GoldCard (GoldCreditCard): Должен содержать дополнительные атрибуты, такие как Бесплатные бонусные мили (BonusMiles). Метод Pay() должен быть переопределен для добавления информации о получении бонусных миль при оплате покупки.
2. PremiumCard (PremiumCreditCard): Должен содержать дополнительные атрибуты, такие как Ассистент поддержки (SupportAssistant). Метод CheckBalance() должен быть переопределен для предоставления возможности обратиться за помощью к ассистенту поддержки в случае проблем с балансом.
3. CorporateCard (CorporateCreditCard) (если требуется третий класс): Должен содержать дополнительные атрибуты, такие как Компания (Company). Метод GetInfo() должен быть переопределен для включения информации о компании в описании карты.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

```
public class CreditCard
{
    public string CardNumber { get; set; }
    public string HolderName { get; set; }
    public string ExpiryDate { get; set; }
    public decimal Balance { get; set; }
    public string CardType { get; set; }

    public CreditCard(string myCardNumber, string myHolderName, string myExpiryDate, decimal myBalance, string myCardType)
    {
        CardNumber = myCardNumber;
        HolderName = myHolderName;
        ExpiryDate = myExpiryDate;
        Balance = myBalance;
        CardType = myCardType;
    }

    public virtual string GetInfo()
    {
        return ($"Информация о кредитной карте: \nНомер карты: {CardNumber}, Владелец: {HolderName}, Срок действия: {ExpiryDate}, Тип карты: {CardType}");
    }

    public virtual void Pay(decimal Minus)
    {
        if (Minus > Balance)
        {
            Console.WriteLine("На вашем счете недостаточно средств");
        }
        else
        {
            Balance -= Minus;
            Console.WriteLine($"Оплата {Minus} рублей успешно произведена. Баланс карты: {Balance}");
        }
    }

    public virtual void CheckBalance()
    {
        Console.WriteLine($"Баланс карты: {Balance}");
    }

    public void Transfer(CreditCard sender, CreditCard recipient, decimal amount)
    {
        if (sender.Balance >= amount)
        {
            sender.Balance -= amount;
            recipient.Balance += amount;

            Console.WriteLine($"Перевод на сумму {amount} рублей со счета карты {sender.CardNumber} на счет карты {recipient.CardNumber} произведен успешно.");
        }
        else
        {
            Console.WriteLine("На счете отправителя недостаточно средств для перевода.");
        }
    }

    public string GetCardType()
    {
        return CardType;
    }
}

public class GoldCard : CreditCard
{
    public decimal BonusMiles { get; set; }
    public int DiscountPercentage { get; set; }

    public GoldCard(string myCardNumber, string myHolderName, string myExpiryDate, decimal myBalance, decimal myBonusMiles, int myDiscountPercentage) : base(myCardNumber, myHolderName, myExpiryDate, myBalance, "Gold")
    {
        BonusMiles = myBonusMiles;
        DiscountPercentage = myDiscountPercentage;
    }

    public override void Pay(decimal Minus)
    {
        base.Pay(Minus);
        int bonusMiles = (int)((int)Minus * 0.1);
        BonusMiles += bonusMiles;
        Console.WriteLine($"Получено {bonusMiles} бонусных миль. Всего бонусных миль: {BonusMiles}");
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Бонусных миль: {BonusMiles}, Скидка: {DiscountPercentage}%";
    }

    public decimal ApplyDiscount(decimal amount)
    {
        return amount - (amount * DiscountPercentage / 100);
    }
}

public class PremiumCard : CreditCard
{
    public string SupportAssistant { get; set; }
    public int CashbackPercentage { get; set; }

    public PremiumCard(string myCardNumber, string myHolderName, string myExpiryDate, decimal myBalance, string mySupportAssistant, int myCashbackPercentage) : base(myCardNumber, myHolderName, myExpiryDate, myBalance, "Premium")
    {
        SupportAssistant = mySupportAssistant;
        CashbackPercentage = myCashbackPercentage;
    }

    public override void CheckBalance()
    {
        base.CheckBalance();
        if (Balance <= 0)
        {
            Console.WriteLine($"Проблема с балансов (на счету недостаточно средств) обратитесь к ассистенту поддержки: {SupportAssistant}.");
        }
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Ассистент поддержки: {SupportAssistant}, Кэшбэк: {CashbackPercentage}%";
    }

    public decimal CalculateCashback(decimal amount)
    {
        return amount * CashbackPercentage / 100;
    }
}

public class CorporateCard : CreditCard
{
    public string Company { get; set; }
    public decimal SpendingLimit { get; set; }

    public CorporateCard(string myCardNumber, string myHolderName, string myExpiryDate, decimal myBalance, string myCompany, decimal mySpendingLimit) : base(myCardNumber, myHolderName, myExpiryDate, myBalance, "Corporate")
    {
        Company = myCompany;
        SpendingLimit = mySpendingLimit;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Компания: {Company}, Лимит расходов: {SpendingLimit}";
    }

    public bool IsSpendingLimitExceeded(decimal amount)
    {
        return amount > SpendingLimit;
    }
}

public class PlatinumCard : GoldCard
{
    public PlatinumCard(string myCardNumber, string myHolderName, string myExpiryDate, decimal myBalance, decimal myBonusMiles, int myDiscountPercentage) : base(myCardNumber, myHolderName, myExpiryDate, myBalance, myBonusMiles, myDiscountPercentage)
    {
    }

    public override string GetInfo()
    {
        return base.GetInfo() + ", Platinum";
    }
}

public interface IHasInsurance
{
    string InsuranceProvider { get; set; }
}

public class InsuranceCard : CreditCard, IHasInsurance
{
    public string InsuranceProvider { get; set; }

    public InsuranceCard(string myCardNumber, string myHolderName, string myExpiryDate, decimal myBalance, string myInsuranceProvider) : base(myCardNumber, myHolderName, myExpiryDate, myBalance, "Insurance")
    {
        InsuranceProvider = myInsuranceProvider;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Страховой провайдер: {InsuranceProvider}";
    }
}

public class SuperCard : GoldCard, IHasInsurance
{
    public string InsuranceProvider { get; set; }

    public SuperCard(string myCardNumber, string myHolderName, string myExpiryDate, decimal myBalance, decimal myBonusMiles, int myDiscountPercentage, string myInsuranceProvider) : base(myCardNumber, myHolderName, myExpiryDate, myBalance, myBonusMiles, myDiscountPercentage)
    {
        InsuranceProvider = myInsuranceProvider;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Страховой провайдер: {InsuranceProvider}";
    }
}


CreditCard myGoldCard = new GoldCard("1111 1111 1111 1111", "Максимка Витаминка", "07/07/2027", 1000000, 777, 10);
CreditCard myPremiumCard = new PremiumCard("3333 3333 3333 3333", "Максимка Витаминка", "08/08/2028", 500000, "Илья Вячеславович", 5);
CreditCard myCorporateCard = new CorporateCard("7777 7777 7777 7777", "Максимка Витаминка", "30/03/2030", 0, "Арбуз.нет", 100000);

CreditCard myPlatinumCard = new PlatinumCard("9999 9999 9999 9999", "Максимка Витаминка", "01/01/2031", 2000000, 1500, 15);

CreditCard myInsuranceCard = new InsuranceCard("5555 5555 5555 5555", "Максимка Витаминка", "09/09/2029", 100000, "Росгосстрах");
CreditCard mySuperCard = new SuperCard("4444 4444 4444 4444", "Максимка Витаминка", "10/10/2030", 1500000, 1000, 12, "АльфаСтрахование");

Console.WriteLine(myGoldCard.GetInfo());
myGoldCard.Pay(10000);

Console.WriteLine();
Console.WriteLine(myPremiumCard.GetInfo());
myPremiumCard.CheckBalance();

Console.WriteLine();
Console.WriteLine(myCorporateCard.GetInfo());

Console.WriteLine();
myGoldCard.Transfer(myGoldCard, myPremiumCard, 50000);
myPremiumCard.Transfer(myPremiumCard, myCorporateCard, 200000);

Console.WriteLine();
Console.WriteLine(myPlatinumCard.GetInfo());

Console.WriteLine();
Console.WriteLine(myInsuranceCard.GetInfo());

Console.WriteLine();
Console.WriteLine(mySuperCard.GetInfo());

Console.WriteLine();
Console.WriteLine($"Тип карты myGoldCard: {myGoldCard.GetCardType()}");
Console.WriteLine($"Тип карты myPremiumCard: {myPremiumCard.GetCardType()}");
Console.WriteLine($"Тип карты myCorporateCard: {myCorporateCard.GetCardType()}");
Console.WriteLine($"Тип карты myPlatinumCard: {myPlatinumCard.GetCardType()}");
Console.WriteLine($"Тип карты myInsuranceCard: {myInsuranceCard.GetCardType()}");
Console.WriteLine($"Тип карты mySuperCard: {mySuperCard.GetCardType()}");

Console.WriteLine();
Console.WriteLine($"Скидка на GoldCard: {((GoldCard)myGoldCard).ApplyDiscount(1000)} рублей");
Console.WriteLine($"Кэшбэк на PremiumCard: {((PremiumCard)myPremiumCard).CalculateCashback(1000)} рублей");
Console.WriteLine($"Превышен ли лимит расходов на CorporateCard: {((CorporateCard)myCorporateCard).IsSpendingLimitExceeded(150000)}");
```
