Описание задачи:

Создать базовый класс CreditCard в C#, который будет представлять информацию о кредитных картах. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма.

Требования к базовому классу CreditCard:

• Атрибуты: Номер карты (CardNumber), Холдера (HolderName), Срок
действия (ExpiryDate).

• Методы:

o GetInfo(): метод для получения информации о кредитной карте в виде
строки.
o Pay(): метод для оплаты покупки с использованием карты.
o CheckBalance(): метод для проверки баланса на карте.

Требования к производным классам:
1. GoldCard (GoldCreditCard): Должен содержать дополнительные атрибуты, такие как Бесплатные бонусные мили (BonusMiles). Метод Pay() должен быть переопределен для добавления информации о получении бонусных миль при оплате покупки.
2. PremiumCard (PremiumCreditCard): Должен содержать дополнительные атрибуты, такие как Ассистент поддержки (SupportAssistant). Метод CheckBalance() должен быть переопределен для предоставления возможности обратиться за помощью к ассистенту поддержки в случае проблем с балансом.
3. CorporateCard (CorporateCreditCard) (если требуется третий класс): Должен содержать дополнительные атрибуты, такие как Компания (Company). Метод GetInfo() должен быть переопределен для включения информации о компании в описании карты.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

```
public class CreditCard
{
    public string CardNumber { get; set; }
    public string HolderName { get; set; }
    public string ExpiryDate { get; set; }
    public decimal Balance { get; set; }
    public bool IsActive { get; set; }
    public int TransactionCount { get; private set; } 

    public CreditCard(string myCardNumber, string myHolderName, string myExpiryDate, decimal myBalance)
    {
        CardNumber = myCardNumber;
        HolderName = myHolderName;
        ExpiryDate = myExpiryDate;
        Balance = myBalance;
        IsActive = true; 
        TransactionCount = 0; 
    }

    public virtual string GetInfo()
    {
        return ($"Информация о кредитной карте: \nНомер карты: {CardNumber}, Владелец: {HolderName}, Срок действия: {ExpiryDate}, Баланс: {Balance}, Статус: {(IsActive ? "Активна" : "Неактивна")}, Количество транзакций: {TransactionCount}");
    }

    public virtual void Pay(decimal Minus)
    {
        if (Minus > Balance)
        {
            Console.WriteLine("На вашем счете недостаточно средств");
        }
        else
        {
            Balance -= Minus;
            TransactionCount++;
            Console.WriteLine($"Оплата {Minus} рублей успешно произведена. Баланс карты: {Balance}");
        }
    }

    public virtual void CheckBalance()
    {
        Console.WriteLine($"Баланс карты: {Balance}");
    }

    public void Transfer(CreditCard sender, CreditCard recipient, decimal amount)
    {
        if (sender != null && recipient != null && sender.Balance >= amount && sender.IsActive && recipient.IsActive) 
        {
            sender.Balance -= amount;
            recipient.Balance += amount;
            sender.TransactionCount++;
            recipient.TransactionCount++; 
            Console.WriteLine($"Перевод на сумму {amount} рублей со счета карты {sender.CardNumber} на счет карты {recipient.CardNumber} произведен успешно.");
        }
        else
        {
            Console.WriteLine("На счете отправителя недостаточно средств для перевода или одна из карт неактивна.");
        }
    }

    public virtual void BlockCard() 
    {
        if (IsActive)
        {
            IsActive = false;
            Console.WriteLine($"Карта {CardNumber} успешно заблокирована.");
        }
        else
        {
            Console.WriteLine($"Карта {CardNumber} уже заблокирована.");
        }
    }
}

public class GoldCard : CreditCard
{
    public decimal BonusMiles { get; set; }
    public int DiscountPercentage { get; set; } 

    public GoldCard(string myCardNumber, string myHolderName, string myExpiryDate, decimal myBalance, decimal myBonusMiles, int myDiscountPercentage) : base(myCardNumber, myHolderName, myExpiryDate, myBalance)
    {
        BonusMiles = myBonusMiles;
        DiscountPercentage = myDiscountPercentage; 
    }

    public override void Pay(decimal Minus)
    {
        base.Pay(Minus);
        int bonusMiles = (int)((int)Minus * 0.1);
        BonusMiles += bonusMiles;
        Console.WriteLine($"Получено {bonusMiles} бонусных миль. Всего бонусных миль: {BonusMiles}");
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Бонусных миль: {BonusMiles}, Скидка: {DiscountPercentage}%";
    }
}

public class PremiumCard : CreditCard
{
    public string SupportAssistant { get; set; }
    public int FreeInsuranceAmount { get; set; } 

    public PremiumCard(string myCardNumber, string myHolderName, string myExpiryDate, decimal myBalance, string mySupportAssistant, int myFreeInsuranceAmount) : base(myCardNumber, myHolderName, myExpiryDate, myBalance)
    {
        SupportAssistant = mySupportAssistant;
        FreeInsuranceAmount = myFreeInsuranceAmount; 
    }

    public override void CheckBalance()
    {
        base.CheckBalance();
        if (Balance <= 0)
        {
            Console.WriteLine($"Проблема с балансом (на счету недостаточно средств) обратитесь к ассистенту поддержки: {SupportAssistant}.");
        }
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Ассистент поддержки: {SupportAssistant}, Бесплатная страховка: {FreeInsuranceAmount} рублей";
    }
}

public class CorporateCard : CreditCard
{
    public string Company { get; set; }
    public decimal CreditLimit { get; set; } 

    public CorporateCard(string myCardNumber, string myHolderName, string myExpiryDate, decimal myBalance, string myCompany, decimal myCreditLimit) : base(myCardNumber, myHolderName, myExpiryDate, myBalance)
    {
        Company = myCompany;
        CreditLimit = myCreditLimit; 
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Компания: {Company}, Кредитный лимит: {CreditLimit} рублей";
    }
}

Console.WriteLine(String.Concat(Enumerable.Repeat("-", 70)));
CreditCard myGoldCard = new GoldCard("1111 1111 1111 1111", "Максимка Витаминка", "07/07/2027", 1000000, 777, 10);
CreditCard myPremiumCard = new PremiumCard("3333 3333 3333 3333", "Максимка Витаминка", "08/08/2028", 500000, "Илья Вячеславович", 100000); 
CreditCard myCorporateCard = new CorporateCard("7777 7777 7777 7777", "Максимка Витаминка", "30/03/2030", 0, "Арбуз.нет", 500000); 

Console.WriteLine(myGoldCard.GetInfo());
myGoldCard.Pay(10000);

Console.WriteLine(String.Concat(Enumerable.Repeat("-", 70)));
Console.WriteLine(myPremiumCard.GetInfo());
myPremiumCard.CheckBalance();

Console.WriteLine(String.Concat(Enumerable.Repeat("-", 70)));
Console.WriteLine(myCorporateCard.GetInfo());

Console.WriteLine(String.Concat(Enumerable.Repeat("-", 70)));
myGoldCard.Transfer(myGoldCard, myPremiumCard, 50000);
myPremiumCard.Transfer(myPremiumCard, myCorporateCard, 200000);

Console.WriteLine(String.Concat(Enumerable.Repeat("-", 70)));
myGoldCard.BlockCard();
myGoldCard.Transfer(myGoldCard, myPremiumCard, 50000);
Console.WriteLine(String.Concat(Enumerable.Repeat("-", 70)));
```
